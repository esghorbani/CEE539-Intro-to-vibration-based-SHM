# Frequency Domain, Laplace Transform, and Proof for a Second-Order System


## 1. Laplace transform
The Laplace transform converts a time-domain signal $x(t)$ into a complex-domain function $X(s)$:

$$
X(s)=\mathcal{L}\{x(t)\}=\int_0^\infty x(t)e^{-st}\,dt
$$

where

$$
s=\sigma + j\omega
$$

- $\sigma$: exponential decay or growth part  
- $\omega$: oscillation part  
- $j=\sqrt{-1}$  

So the Laplace domain is more general than the frequency domain.

---

### Frequency domain
The **frequency domain** describes how a system behaves for sinusoidal inputs at different frequencies.

To get the frequency response from the Laplace domain, we evaluate the transfer function on the imaginary axis:

$$
s=j\omega
$$

So if the transfer function is $H(s)$, then the frequency response is

$$
H(j\omega)
$$

This tells us:

- amplitude change
- phase shift

at frequency $\omega$.

---

## 2. Why $s=j\omega$?

A pure sinusoid is related to the complex exponential:

$$
e^{j\omega t}=\cos(\omega t)+j\sin(\omega t)
$$

A sinusoidal steady-state signal has no exponential growth or decay, so $\sigma=0$. Therefore,

$$
s=\sigma+j\omega \quad \Rightarrow \quad s=j\omega
$$

That is why frequency-domain analysis uses $s=j\omega$.

---

## 3. Laplace property for derivatives

This is the main idea behind the proof.

If

$$
X(s)=\mathcal{L}\{x(t)\}
$$

then

$$
\mathcal{L}\{\dot{x}(t)\}=sX(s)-x(0)
$$

and

$$
\mathcal{L}\{\ddot{x}(t)\}=s^2X(s)-sx(0)-\dot{x}(0)
$$

For **zero initial conditions**,

$$
x(0)=0,\qquad \dot{x}(0)=0
$$

so

$$
\mathcal{L}\{\dot{x}(t)\}=sX(s)
$$

$$
\mathcal{L}\{\ddot{x}(t)\}=s^2X(s)
$$

Now in the frequency domain, substitute $s=j\omega$:

$$
\dot{x}(t) \leftrightarrow j\omega X(j\omega)
$$

and

$$
\ddot{x}(t) \leftrightarrow (j\omega)^2 X(j\omega)=-\omega^2 X(j\omega)
$$

So:

- first derivative becomes $j\omega$
- second derivative becomes $-\omega^2$

---

## 4. Proof for the second-order system

Consider the standard second-order differential equation:

$$
\ddot{x}(t)+2\zeta\omega_n\dot{x}(t)+\omega_n^2 x(t)=f(t)
$$

Take the Laplace transform with zero initial conditions:

$$
s^2X(s)+2\zeta\omega_n sX(s)+\omega_n^2X(s)=F(s)
$$

Factor out $X(s)$:

$$
\left(s^2+2\zeta\omega_n s+\omega_n^2\right)X(s)=F(s)
$$

Therefore, the transfer function is

$$
H(s)=\frac{X(s)}{F(s)}=\frac{1}{s^2+2\zeta\omega_n s+\omega_n^2}
$$

---

## 5. Move from Laplace domain to frequency domain

To get the frequency response, set

$$
s=j\omega
$$

Then

$$
H(j\omega)=\frac{1}{(j\omega)^2+2\zeta\omega_n(j\omega)+\omega_n^2}
$$

Now simplify each term.

### First term
$$
(j\omega)^2=-\omega^2
$$

### Second term
$$
2\zeta\omega_n(j\omega)=j\,2\zeta\omega_n\omega
$$

### Third term
$$
\omega_n^2=\omega_n^2
$$

So the denominator becomes

$$
-\omega^2 + j\,2\zeta\omega_n\omega + \omega_n^2
$$

Rearranging:

$$
\omega_n^2-\omega^2+j\,2\zeta\omega_n\omega
$$

Thus, the frequency response is

$$
H(j\omega)=\frac{1}{\omega_n^2-\omega^2+j\,2\zeta\omega_n\omega}
$$

---

## 6. Equivalent proof directly in frequency domain

Starting from

$$
\ddot{x}(t)+2\zeta\omega_n\dot{x}(t)+\omega_n^2x(t)=f(t)
$$

Use the frequency-domain substitutions:

$$
\ddot{x}(t)\rightarrow -\omega^2 X(j\omega)
$$

$$
\dot{x}(t)\rightarrow j\omega X(j\omega)
$$

$$
x(t)\rightarrow X(j\omega)
$$

Then the equation becomes

$$
-\omega^2X(j\omega)+j\,2\zeta\omega_n\omega X(j\omega)+\omega_n^2X(j\omega)=F(j\omega)
$$

Factor out $X(j\omega)$:

$$
\left(\omega_n^2-\omega^2+j\,2\zeta\omega_n\omega\right)X(j\omega)=F(j\omega)
$$

So,

$$
\frac{X(j\omega)}{F(j\omega)}
=
\frac{1}{\omega_n^2-\omega^2+j\,2\zeta\omega_n\omega}
$$

This is exactly the desired result.

---

## 7. Physical meaning

The denominator

$$
\omega_n^2-\omega^2+j\,2\zeta\omega_n\omega
$$

has:

- **real part**:

$$
\omega_n^2-\omega^2
$$

- **imaginary part**:

$$
2\zeta\omega_n\omega
$$

Interpretation:

- the real part is related to stiffness and inertia
- the imaginary part is related to damping

So the frequency-domain form comes directly from replacing derivatives by algebraic terms.

---

## 8. Final compact result

For the second-order system

$$
\ddot{x}(t)+2\zeta\omega_n\dot{x}(t)+\omega_n^2x(t)=f(t)
$$

the transfer function in the Laplace domain is

$$
H(s)=\frac{1}{s^2+2\zeta\omega_n s+\omega_n^2}
$$

and the frequency response is

$$
H(j\omega)=\frac{1}{\omega_n^2-\omega^2+j\,2\zeta\omega_n\omega}
$$

This comes from the rules:

$$
\dot{x}(t)\rightarrow j\omega X(j\omega)
$$

$$
\ddot{x}(t)\rightarrow -\omega^2 X(j\omega)
$$

# FFT of a Signal and How to Plot It

## 1. Goal

The Fast Fourier Transform (FFT) converts a signal from the **time domain** into the **frequency domain**.

If a sampled signal is

$$
x[0], x[1], x[2], \dots, x[N-1]
$$

the FFT reveals the **frequency components** contained in the signal.

---

## 2. Sampling the Signal

A continuous signal $x(t)$ must first be sampled.

Let

$$
f_s
$$

be the **sampling frequency (Hz)**.

The sampling interval is

$$
\Delta t = \frac{1}{f_s}
$$

The sampled signal becomes

$$
x[n] = x(n\Delta t), \quad n = 0,1,\dots,N-1
$$

where $N$ is the total number of samples.

The total duration of the signal is

$$
T = N \Delta t = \frac{N}{f_s}
$$

---

## 3. Nyquist Frequency

The highest frequency that can be correctly represented is the **Nyquist frequency**

$$
f_{Nyquist} = \frac{f_s}{2}
$$

Any frequency higher than this causes **aliasing**.

Therefore, when plotting the FFT of a real-valued signal, we normally display only

$$
0 \le f \le \frac{f_s}{2}
$$

because the spectrum is symmetric.

---

## 4. Number of Samples

The number of samples in the signal is

$$
N = \text{length}(x)
$$

---

## 5. Choose the FFT Length

The FFT can be computed using $N$ samples, but it is common to use the **next power of 2** for computational efficiency.

The FFT length becomes

$$
N_{FFT} = 2^{\lceil \log_2(N) \rceil}
$$

If

$$
N_{FFT} > N
$$

then the signal is **zero padded**.

---

## 6. Compute the FFT

Use the FFT function to compute the transform of the signal:

$$
X = \mathrm{fft}(x, N_{FFT})
$$

---

## 7. Frequency Axis

Each FFT index corresponds to a frequency

$$
f_k = \frac{k f_s}{N_{FFT}}
$$

for

$$
k = 0,1,\dots,N_{FFT}-1
$$

---

## 8. Plot the FFT

To plot the FFT:

- use the magnitude of the FFT, $|X[k]|$
- plot the spectrum only up to half of the sampling frequency:

$$
0 \le f \le \frac{f_s}{2}
$$

This is the usual plotting range for a real-valued signal.

---

## 9. Log Scale for Better Peak Visibility

Sometimes, to see peaks more clearly, especially small peaks next to large ones, the amplitude axis is shown in **log scale**.

So the FFT can be plotted:

- either in normal scale
- or with the **y-axis in logarithmic scale** for better visibility of peaks

---

## 10. Summary of the Procedure

1. Sample the signal $x(t)$ using sampling frequency $f_s$.
2. Obtain the discrete signal $x[n]$.
3. Determine the number of samples $N$.
4. Choose FFT length $N_{FFT}$, often as the next power of 2.
5. Use the FFT function to compute

$$
X = \mathrm{fft}(x, N_{FFT})
$$

6. Build the frequency axis.
7. Plot the FFT magnitude only up to

$$
\frac{f_s}{2}
$$

8. If needed, use a logarithmic y-axis to make peaks easier to distinguish.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eig
from scipy.integrate import solve_ivp

# ----------------------------
# 1) Define the 3DOF system
# ----------------------------
m = 1.0
k = 1.0

M = np.array([
    [2*m, 0,   0],
    [0,   3*m, 0],
    [0,   0,   1*m]
], dtype=float)

K = np.array([
    [ 5*k, -3*k,  0   ],
    [-3*k,  5*k, -2*k ],
    [ 0,   -2*k,  3*k ]
], dtype=float)

# Light proportional damping
alpha = 0.0
beta = 0.0
C = alpha * M + beta * K

# ----------------------------
# 2) State-space model
# ----------------------------
Minv = np.linalg.inv(M)

A_top = np.hstack((np.zeros((3, 3)), np.eye(3)))
A_bottom = np.hstack((-Minv @ K, -Minv @ C))
A = np.vstack((A_top, A_bottom))

def state_eq(t, z):
    return A @ z

# Initial condition
x0 = np.array([1.0, 0.5, -0.2])
v0 = np.array([0.0, 0.0, 0.0])
z0 = np.hstack((x0, v0))

dt = 0.01
t_end = 60.0
t_eval = np.arange(0, t_end, dt)

sol = solve_ivp(state_eq, [0, t_end], z0, t_eval=t_eval, method='RK45')

# Displacement and velocity from state solution
x = sol.y[:3, :]     # shape (3, N)
xdot = sol.y[3:, :]  # shape (3, N)
t = sol.t
N = len(t)
fs = 1.0 / dt

# Acceleration from equation of motion:
# M xddot + C xdot + K x = 0   ->   xddot = -M^{-1}(C xdot + K x)
xddot = -(Minv @ (C @ xdot + K @ x))

# ----------------------------
# 3) ITD estimation (kept from your code)
# ----------------------------
X = x
X1 = np.vstack((X[:, 0:N-2], X[:, 1:N-1]))   # shape (6, N-2)
X2 = np.vstack((X[:, 1:N-1], X[:, 2:N]))     # shape (6, N-2)

A_d = X2 @ np.linalg.pinv(X1)

mu, V = eig(A_d)
s = np.log(mu) / dt

selected = np.array([pole for pole in s if np.imag(pole) > 1e-6])
selected = selected[np.argsort(np.imag(selected))]

omega_d = np.abs(np.imag(selected))
omega_n = np.sqrt(np.real(selected)**2 + np.imag(selected)**2)
zeta = -np.real(selected) / omega_n
f_n = omega_n / (2 * np.pi)

print("Estimated poles:")
print(selected)

print("\nEstimated damped natural frequencies ω_d [rad/s]:")
print(omega_d)

print("\nEstimated natural frequencies ω_n [rad/s]:")
print(omega_n)

print("\nEstimated natural frequencies f_n [Hz]:")
print(f_n)

print("\nEstimated damping ratios ζ:")
print(zeta)

# ----------------------------
# 4) FFT settings
# ----------------------------
f_plot_max = 2.0      # Hz -> change this to zoom in/out on x-axis
use_log_y = False      # True if you want log scale on amplitude axis

# ----------------------------
# 5) FFT helper
# ----------------------------
def single_sided_fft(signal, fs):
    """
    Returns:
        f : frequency axis from 0 to fs/2
        A : single-sided amplitude spectrum
    """
    N = len(signal)
    NFFT = 2**int(np.ceil(np.log2(N)))   # next power of 2

    Y = np.fft.fft(signal, n=NFFT)
    f = np.fft.rfftfreq(NFFT, d=1/fs)

    A = np.abs(Y[:len(f)]) / N
    if len(A) > 2:
        A[1:-1] = 2.0 * A[1:-1]

    return f, A

# ----------------------------
# 6) Plot FFTs in 3x3 layout
# Rows = DOFs
# Cols = x, xdot, xddot
# ----------------------------
responses = [x, xdot, xddot]
col_titles = [r'FFT of $x$', r'FFT of $\dot{x}$', r'FFT of $\ddot{x}$']

fig, axes = plt.subplots(3, 3, figsize=(16, 10), constrained_layout=True)

for dof in range(3):
    for j, resp in enumerate(responses):
        f_fft, A_fft = single_sided_fft(resp[dof, :], fs)

        ax = axes[dof, j]
        ax.plot(f_fft, A_fft, lw=1.2)
        ax.set_xlim(0, f_plot_max)
        ax.grid(True, alpha=0.3)

        if use_log_y:
            ax.set_yscale('log')

        if dof == 0:
            ax.set_title(col_titles[j], fontsize=12)

        if j == 0:
            ax.set_ylabel(f'DOF {dof+1}\nAmplitude')
        else:
            ax.set_ylabel('Amplitude')

        ax.set_xlabel('Frequency [Hz]')

        # Optional: mark identified frequencies
        for fn in f_n:
            if np.isfinite(fn) and fn > 0 and fn <= f_plot_max:
                ax.axvline(fn, color='r', linestyle='--', alpha=0.35)

plt.suptitle('Single-Sided FFT of x, xdot, and xddot for Each DOF', fontsize=14)
plt.show()
```

# Peak Picking Method


The **peak picking method** is a simple frequency-domain method used in vibration and modal analysis to estimate the **natural frequencies** of a system from its response signal.

The idea is straightforward:

- compute the FFT of the vibration response
- plot amplitude versus frequency
- identify the main peaks in the spectrum

These peaks usually occur near the **natural frequencies** of the system.

This method works best for:

- lightly damped systems
- well-separated modes
- clean response signals

---

## 2. How to Find the Natural Frequency

After computing the FFT of the signal, look at the frequency spectrum.

If the amplitude spectrum has a peak at frequency

$$
f_{peak}
$$

then the natural frequency is approximated by $
f_n \approx f_{peak}
$

So, in the peak picking method:

- each major peak in the FFT corresponds to a vibration mode
- the frequency location of that peak is taken as the natural frequency

In short:

$$
\text{natural frequency} = \text{frequency at the spectral peak}
$$

---

## 3. How to Estimate the Damping Ratio

A simple rule for estimating damping ratio from the spectrum is the **half-power bandwidth method**.

First, find the main resonance peak at

$$
f_n
$$

Let the amplitude at the peak be

$$
A_{peak}
$$

Then find two frequencies on both sides of the peak:

$f_1$ = left frequency where the amplitude drops to

$$
\frac{A_{peak}}{\sqrt{2}}
$$

 $f_2$ = right frequency where the amplitude drops to

$$
\frac{A_{peak}}{\sqrt{2}}
$$

These are called the **half-power points**.

The damping ratio is then estimated by

$$
\zeta \approx \frac{f_2 - f_1}{2f_n}
$$

where:

 $f_n$ = natural frequency and 
 $f_2 - f_1$ = bandwidth around the peak

---

## 4. Interpretation

- a **sharp and narrow peak** means small damping
- a **wide peak** means larger damping

So:

- natural frequency comes from the **location of the peak**
- damping ratio comes from the **width of the peak**

---

## 5. Summary

### Natural frequency

Find the FFT peak and use its frequency:

$$
f_n \approx f_{peak}
$$

### Damping ratio

Use the half-power bandwidth rule:

$$
\zeta \approx \frac{f_2 - f_1}{2f_n}
$$

This method is simple and useful, but it is most reliable for **lightly damped systems with clearly separated peaks**.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# -----------------------------------
# System parameters
# -----------------------------------
m = 1.0
k = 90.0
c = 4.0          # increased damping to widen the peak

F0 = 0.0
omega = 3.0

# -----------------------------------
# Time settings
# -----------------------------------
dt = 0.001
t_end = 10.0
n = int(t_end / dt)

# -----------------------------------
# Initial conditions
# -----------------------------------
x0 = 20.0
v0 = 0.0

# -----------------------------------
# Arrays
# -----------------------------------
t = np.zeros(n + 1)
x = np.zeros(n + 1)
v = np.zeros(n + 1)
a = np.zeros(n + 1)
f = np.zeros(n + 1)

# Initial values
x[0] = x0
v[0] = v0
f[0] = F0 * np.sin(omega * t[0])
a[0] = (f[0] - c * v[0] - k * x[0]) / m

# -----------------------------------
# Euler integration
# -----------------------------------
for i in range(n):
    t[i + 1] = t[i] + dt
    f[i] = F0 * np.sin(omega * t[i])
    a[i] = (f[i] - c * v[i] - k * x[i]) / m
    x[i + 1] = x[i] + dt * v[i]
    v[i + 1] = v[i] + dt * a[i]

f[n] = F0 * np.sin(omega * t[n])
a[n] = (f[n] - c * v[n] - k * x[n]) / m

# -----------------------------------
# FFT of displacement
# -----------------------------------
fs = 1.0 / dt
N = len(x)
NFFT = 2**int(np.ceil(np.log2(N)))

X = np.fft.fft(x, NFFT)
freq = np.fft.rfftfreq(NFFT, d=dt)

A = np.abs(X[:len(freq)]) / N
if len(A) > 2:
    A[1:-1] *= 2.0

# -----------------------------------
# Peak picking
# -----------------------------------
# Ignore DC and very low frequency bins
valid = freq > 0.05
freq_valid = freq[valid]
A_valid = A[valid]

peaks, _ = find_peaks(A_valid, prominence=np.max(A_valid) * 0.05)

if len(peaks) == 0:
    raise ValueError("No peak found. Try increasing simulation time or adjusting damping.")

# Main peak
main_peak_idx_local = peaks[np.argmax(A_valid[peaks])]
main_peak_idx = np.where(valid)[0][main_peak_idx_local]

f_n = freq[main_peak_idx]
A_peak = A[main_peak_idx]

# Half-power level
A_half = A_peak / np.sqrt(2)

# -----------------------------------
# Find left half-power frequency f1
# -----------------------------------
left_idx = main_peak_idx
while left_idx > 1 and A[left_idx] > A_half:
    left_idx -= 1

if left_idx == 1:
    raise ValueError("Left half-power point not found.")

# Linear interpolation for better estimate
f1 = np.interp(
    A_half,
    [A[left_idx], A[left_idx + 1]],
    [freq[left_idx], freq[left_idx + 1]]
)

# -----------------------------------
# Find right half-power frequency f2
# -----------------------------------
right_idx = main_peak_idx
while right_idx < len(A) - 2 and A[right_idx] > A_half:
    right_idx += 1

if right_idx == len(A) - 2:
    raise ValueError("Right half-power point not found.")

# Linear interpolation for better estimate
f2 = np.interp(
    A_half,
    [A[right_idx], A[right_idx - 1]],
    [freq[right_idx], freq[right_idx - 1]]
)

# -----------------------------------
# Damping ratio from half-power bandwidth
# -----------------------------------
zeta_est = (f2 - f1) / (2.0 * f_n)

print(f"Peak frequency f_n = {f_n:.4f} Hz")
print(f"Half-power frequencies: f1 = {f1:.4f} Hz, f2 = {f2:.4f} Hz")
print(f"Estimated damping ratio zeta = {zeta_est:.4f}")

# Theoretical values for comparison
omega_n_theory = np.sqrt(k / m)
f_n_theory = omega_n_theory / (2 * np.pi)
zeta_theory = c / (2 * np.sqrt(m * k))

print(f"Theoretical natural frequency = {f_n_theory:.4f} Hz")
print(f"Theoretical damping ratio = {zeta_theory:.4f}")

# -----------------------------------
# Plot time response
# -----------------------------------
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

axes[0].plot(t, x)
axes[0].set_title('Displacement x(t)')
axes[0].set_xlabel('Time (s)')
axes[0].set_ylabel('x')
axes[0].grid(True)

# -----------------------------------
# Plot FFT with peak picking
# -----------------------------------
axes[1].plot(freq, A, label='FFT amplitude')
axes[1].axvline(f_n, color='k', linestyle='--', label=f'Peak: {f_n:.3f} Hz')
axes[1].axvline(f1, color='r', linestyle='--', label=f'f1 = {f1:.3f} Hz')
axes[1].axvline(f2, color='r', linestyle='--', label=f'f2 = {f2:.3f} Hz')
axes[1].axhline(A_half, color='g', linestyle=':', label=r'Half-power level')

axes[1].set_xlim(0, 5)
axes[1].set_title('FFT and Peak Picking')
axes[1].set_xlabel('Frequency (Hz)')
axes[1].set_ylabel('Amplitude')
axes[1].grid(True)
axes[1].legend()

plt.tight_layout()
plt.show()
```
